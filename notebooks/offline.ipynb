```python
import numpy as np
import pandas as pd

from ensalada.session import Session
from ensalada.model import LabeledLDA

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='ticks', context='talk')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.sans-serif'] = "Arial"
```

```python
sess = Session("AJF027", "CDEF", 2, bin_size=0.1)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111)
plt.plot(*sess.position().values.T, color='grey', alpha=0.5, linewidth=1)
ax.add_patch(plt.Rectangle((-50, -50), 100, 100, facecolor='red', alpha=0.1))
plt.axis('square')
plt.savefig('position_tracking.pdf')
```

```python
X = sess.format_offline_data()
```

```python
plt.figure(figsize=(6, 3), dpi=100)
sns.heatmap(X, cmap='rainbow', yticklabels=500, xticklabels=20, cbar_kws={'label': 'Spike count'}, rasterized=True)
plt.ylabel('Samples')
plt.xlabel('Neurons')
plt.yticks(rotation=None)

sns.despine()
plt.tight_layout()

plt.savefig('raw_data_offline.pdf')
```
