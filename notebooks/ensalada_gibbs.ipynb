```python
from ensalada.session import Session
from ensalada.llda import LldaModel
from ensalada.utils import to_trial_labels, raster2documents, zip_rasters_labels
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
```

```python
sess = Session("AJF023", "CDEF", 3)
```

```python
ras = sess.rasters()
```

```python
plt.imshow(np.hstack(ras), aspect='auto', vmin=0, vmax=1)
plt.colorbar()
```

```python
tl = sess.labels()
```

```python
all_labeled_documents = zip_rasters_labels(ras, list(tl))
```

```python
model = LldaModel(labeled_documents=all_labeled_documents)
```
